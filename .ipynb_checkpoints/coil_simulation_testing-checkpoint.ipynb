```python
import numpy as np
import matplotlib.pyplot as plt

# Physical constant
mu_0 = 4 * np.pi * 1e-7  # Vacuum permeability (T·m/A)

def helmholtz_magnetic_field(N, R, I, z_range):
    """
    Compute the magnetic field along the z-axis for a Helmholtz coil.

    Parameters:
    - N: number of turns in each coil
    - R: radius of the coils (meters)
    - I: current in the coils (Amperes)
    - z_range: numpy array of z positions (meters)

    Returns:
    - Bz: magnetic field at each z in Tesla
    """
    # Distance between the coils = R for Helmholtz configuration
    d = R / 2  # distance from the center to each coil

    # Calculate B-field from both coils at each z position
    def single_loop_field(z, z0):
        return (mu_0 * N * I * R**2) / (2 * ((R**2 + (z - z0)**2)**(1.5)))

    Bz = single_loop_field(z_range, -d) + single_loop_field(z_range, +d)
    return Bz/10000

# Example usage
if __name__ == "__main__":
    # Input parameters
    N = 100        # Number of turns
    R = 0.1        # Radius in meters (10 cm)
    I = 1.0        # Current in Amperes
    z = np.linspace(-0.2, 0.2, 500)  # z from -20 cm to 20 cm

    # Calculate magnetic field
    Bz = helmholtz_magnetic_field(N, R, I, z)

    # Plotting
    plt.figure(figsize=(8, 5))
    plt.plot(z, Bz, label='Magnetic Field Bz (Gauss)')
    plt.xlabel('z (m)')
    plt.ylabel('Bz (G)')
    plt.title('Magnetic Field Along z-axis of a Helmholtz Coil')
    plt.grid(True)
    plt.legend()
    plt.show()
```

```python


# Physical constant
mu_0 = 4 * np.pi * 1e-7  # Vacuum permeability (T·m/A)

def single_loop_field(z, z0, R, I, N):
    """
    Magnetic field along the z-axis from a single circular current loop.

    Parameters:
    - z: numpy array of positions along the z-axis
    - z0: z-position of the center of the loop
    - R: radius of the loop (meters)
    - I: current in the loop (Amperes)
    - N: number of turns in the loop

    Returns:
    - Bz: magnetic field at each z position (Tesla)
    """
    return (mu_0 * N * I * R**2) / (2 * ((R**2 + (z - z0)**2)**(1.5)))

def total_magnetic_field(z, coil_configs):
    """
    Compute total magnetic field along the z-axis from multiple coils.

    Parameters:
    - z: numpy array of z positions
    - coil_configs: list of dictionaries, each with keys:
        {'z0': float, 'R': float, 'I': float, 'N': int}

    Returns:
    - Bz_total: total magnetic field along z-axis (Tesla)
    """
    Bz_total = np.zeros_like(z)
    for coil in coil_configs:
        Bz_total += single_loop_field(
            z=z,
            z0=coil['z0'],
            R=coil['R'],
            I=coil['I'],
            N=coil['N']
        )
    return Bz_total*10000

# Example usage
if __name__ == "__main__":
    # z positions along the axis
    z = np.linspace(-0.8, 0.8, 1000)  # meters

    # Define coils (Helmholtz configuration + 2 extra coils)
    coil_configs = [
        {'z0': -0.4074, 'R': 0.435, 'I': -10, 'N': 84},  # Coil 1
        {'z0': -0.4989, 'R': 0.435, 'I': -10, 'N': 84},  # Coil 2 (Helmholtz pair)
        {'z0': 0.4074, 'R': 0.435, 'I': -10, 'N': 84},   # Extra Coil 3
        {'z0':  0.4984, 'R': 0.435, 'I': -10, 'N': 84}    # Extra Coil 4
    ]

    # Compute total field
    Bz = total_magnetic_field(z, coil_configs)

    # Plotting
    plt.figure(figsize=(9, 5))
    plt.plot(z, Bz, label='Total Magnetic Field Bz (Gauss)')
    plt.xlabel('z (m)')
    plt.ylabel('Bz (G)')
    plt.title('Magnetic Field Along z-axis from Multiple Coils')
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()
```
